### ASSIGNMENT 1: 

```python
import numpy as np
from pathlib import Path
from PIL import Image

train = Path("C:/Users/USER/Desktop/DS102-Học máy thống kê/BTH3/archive/chest_xray/train")
test = Path("C:/Users/USER/Desktop/DS102-Học máy thống kê/BTH3/archive/chest_xray/test")

train_path = list(train.rglob("*.jpeg"))
test_path = list(test.rglob("*.jpeg"))

x_train = []
y_train = []
x_test = []
y_test = []

label_maps = {
    "NORMAL": -1,
    "PNEUMONIA": 1
}

for path in train_path:
    img = Image.open(path).convert("L")
    img = img.resize((128, 128))

    img_array = np.array(img)
    label = label_maps[path.parent.name]

    img_flat = img_array.flatten()
    img_normalize = img_flat / 255.0

    x_train.append(img_normalize)
    y_train.append(label)
 

for path in test_path:
    img = Image.open(path).convert("L")
    img = img.resize((128,128))
    img_array = np.array(img)

    img_flat = img_array.flatten()
    img_normalize = img_flat / 255.0

    label = label_maps[path.parent.name]
    y_test.append(label)
    x_test.append(img_normalize)

x_train = np.array(x_train)

```

```python
w = np.zeros(128*128)
b = 0.0

epochs = 100
learning_rate = 0.001
C = 1.0

for epoch in range(epochs):
    for i in range(len(x_train)):
        xi = x_train[i]
        yi = y_train[i]

        S = np.dot(w, xi) + b
        if yi*S >= 1:
            w = w - learning_rate *(w/ len(x_train))
        else:
            w = w - learning_rate*((w/len(x_train)) -C*yi*xi)
            b = b - learning_rate*(-C*yi)
```

```python
y_pred = []

for i in range (len(x_test)):
    xi = x_test[i]
    S = np.dot(w, xi)+b

    if S >= 0:
        y_pred.append(1)
    else:
        y_pred.append(-1)

TP = 0
TN = 0
FP = 0
FN = 0
for i in range(len(y_test)):
    if(y_test[i]==1 and y_pred[i]==1):
        TP+=1
    elif(y_test[i]==-1 and y_pred[i]==-1):
        TN+=1
    elif(y_test[i]==1 and y_pred[i]==-1):
        FN+=1
    else:
        FP+=1

Precision = TP/(TP+FP)
Recall = TP/(TP+FN)
F1_score = 2*(Precision*Recall)/(Precision+Recall)

print("Precision:", Precision)
print("Recall:", Recall)   
print("F1-score:", F1_score)
```

### ASSIGNMENT 2:

```python
from sklearn.svm import SVC

model = SVC(kernel = 'linear', C = 1.0)

model.fit(x_train, y_train)

y_pred1 = model.predict(x_test)

```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test, y_pred1, pos_label = 1)
recall = recall_score(y_test, y_pred1, pos_label = 1)
f1 = f1_score(y_test, y_pred1, pos_label = 1)

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
